```python
# dependencies and setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from sqlalchemy import create_engine
from passwords import password
```

```python
# import the SQL database into Pandas:
# create an engine that can talk to the database (password replaced with string PASSWORD)
engine = create_engine(f"postgresql+psycopg2://postgres:{password}@localhost:5432/ETL_Project")
conn = engine.connect()
```

```python
avg_Trump = pd.read_sql(
    "SELECT p.pollster, AVG(p.adjpoll_trump), AVG(d.pct) FROM data_2016 as p JOIN data_2020 as d ON p.pollster = d.pollster WHERE d.population = 'lv' AND p.population = 'lv' AND d.candidate_name = 'Donald Trump' GROUP BY p.pollster;",
    conn)
```

```python
# rename columns and preview the data
avg_Trump.columns = ["pollster", "2016_avg", "2020_avg"]
avg_Trump
```

```python
# create a bar chart displaying the data side by side
pollsters = avg_Trump["pollster"].tolist()
avg_2016 = round(avg_Trump["2016_avg"], 1).tolist()
avg_2020 = round(avg_Trump["2020_avg"], 1).tolist()

# copy/pasted format for my chart from https://matplotlib.org/3.1.1/gallery/lines_bars_and_markers/barchart.html

x = np.arange(len(pollsters))  # the label locations
width = 0.4  # the width of the bars

fig, ax = plt.subplots(figsize=(9,8))
rects1 = ax.bar(x - width/2, avg_2016, width, color="red", alpha=0.6, label='Trump 2016')
rects2 = ax.bar(x + width/2, avg_2020, width, color="blue", alpha=0.6, label='Trump 2020')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel("Average Poll %", fontsize=14)
ax.set_title("Polling for Trump 2016, 2020", fontsize=18, fontweight="bold")
ax.set_xticks(x)
ax.set_xticklabels(pollsters, rotation=90, fontsize=12)
ax.legend()

def autolabel(rects):
    # Attach a text label above each bar in *rects*, displaying its height
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 2),  # 2 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()
plt.hlines(50, -0.5, (len(x)-0.5), colors="grey", linestyles="dashed", alpha=0.5)
plt.savefig("Polling_Trump.png", dpi=300)
plt.show()
```
