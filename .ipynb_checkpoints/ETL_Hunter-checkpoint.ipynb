## 2016 Presidential Polls

```python
# read in the csv file
polls_2016 = pd.read_csv("Resources/presidential_polls.csv")
```

```python
# missing values table function I found on kaggle

def missing_values_table(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum() / len(df)
    mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values'})
    mis_val_tables_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] !=0].sort_values(
    '% of Total Values', ascending=False).round(1)
    print("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"
         "There are " + str(mis_val_table_ren_columns.shape[0]) +
         " columns that have missing values.")
    return mis_val_table_ren_columns    
```

```python
# how many different pollsters are there here
len(polls_2016["pollster"].unique())
```

```python
# does each column have the same number of data elements
polls_2016.count()
```

```python
# missing values table function
missing_values_table(polls_2016)
```

```python
new_df  = polls_2016[["url", "state", "population"]]
```

## 2020 Presidential Polls - multiple sources aggregated

```python
# read in the csv file
polls_2020 = pd.read_csv("Resources/president_polls_2.csv")
```

```python
# how many different pollsters are there here
len(polls_2020["pollster"].unique())
```

```python
list = polls_2020["pollster"].unique().tolist()
list
```

```python
list2 = polls_2016["pollster"].unique().tolist()
list2
```

```python
# return a list of pollsters feature in 2016 and reappear in 2020
y = []
for x in list:
    if x in list2:
        y.append(x)
print(y)
print(len(y))
```

```python
# missing values table function
missing_values_table(polls_2020)
```

```python
# check count of each population type
polls_2020["population_full"].value_counts()
```

```python
# how many elements for each column
new_2020.count()
```

```python
# suspect the missing state values belong to specific pollsters, not randomly distributed
new_2020.loc[new_2020["pollster"] == "Morning Consult"].count()
```

```python
# create df with only the pollsters in common with their state data
column_names = ["state", "pollster"]

r_df = pd.DataFrame(columns = column_names)

for a in y:
    df = new_2020.loc[new_2020["pollster"] == a]
    r_df = r_df.append(df)
r_df
```
